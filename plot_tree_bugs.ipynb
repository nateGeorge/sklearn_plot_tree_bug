```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('prepped_diabetes_data.csv', index_col='Patient number')
df
```

```python
features = df.drop('Diabetes', axis=1)
targets = df['Diabetes']
```

```python
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(features, targets)
```

```python
# saved figure is squished to the bottom left if dpi is increased
f = plt.figure(figsize=(10, 10))
plot_tree(dt, fontsize=8, feature_names=features.columns, filled=True)
plt.savefig('decision_tree.jpg', dpi=300)
plt.show()
```
